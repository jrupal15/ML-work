```python
import matplotlib.pyplot as p
x=[2,5]
y=[6,9]
x1=[22,34]
y1=[12,30]

p.xlabel("marks")
p.ylabel("ranking")
p.plot(x,y)
p.plot(x1,y1)
p.show()
```

```python
import matplotlib.pyplot as p
x=[2,5]
y=[6,9]
x1=[4,2]
y1=[1,3]

p.xlabel("marks")
p.ylabel("ranking")
p.plot(x,y,label='subject1')
p.plot(x1,y1,label="subject2")
p.legend()
p.show()

```
